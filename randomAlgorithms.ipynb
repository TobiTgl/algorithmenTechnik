function RandomIncrementalCH()
    Construct the convex hull CH of P1, P2, P3 in clockwise order, stored in a doubly-linked list.
    
    Compute a point C inside the convex hull (e.g., the centroid (P1 + P2 + P3)/3).
    
    Randomly permute the remaining points, and call the new order P4, P5, . . . , Pn.
    
    For each P4, . . . , Pn, compute the edge of CH intersected by ray ‚àí‚àí‚ÜíCPi, and associate this edge with Pi.
    
    For i = 4, . . . , n:
        Retrieve the associated edge e of Pi, which is visible from Pi.
        
        Compute the intersection of ‚àí‚àí‚ÜíCP and e; call this point Q. If length CQ is greater than length CP, then P is inside CH, so do nothing and continue onto the next iteration.

        Run BuildTent(CH, P, e).

        For each deleted edge, reassign the future points associated with that edge to whichever of LP and PR that intersects ray ‚àí‚àí‚ÜíCPi.

function BuildTent(CH, P, e)
    Starting from e, go counterclockwise along CH until the next edge to be visited is no longer visible from P. Let L be the left endpoint of the last visible edge.
    
    Do the same in the clockwise direction starting from e. Let R be the right endpoint of
    the last visible edge.

    Remove all visited edges in both directions, and add the edges LP and PR in their place in the doubly-linked list.

Fermat Primality Test (Randomized)

Input: n (the number to test for primality), k (the number of iterations)
Output: "probably prime" or "composite"

function fermatPrimalityTest(n, k):
    if n <= 1:
        return "n is not a prime/composite"
    if n <= 3:
        return "probably prime"
    if n % 2 == 0:
        return "composite"

    for i = 1 to k:
        a = random(2, n - 2)  // Pick a random integer in the range [2, n-2] # ab 2 da 1 als basis kein sinn macht 
        if power(a, n - 1) mod n != 1 oder power(a, n - 1, n) != 1:
            return "composite"
        
    return "probably prime"


function power(a, b, n)
    if b==0
        return 1
    else if b mod 2 == 0 # even
        d = power(a, b/2, n)
        return d*d mod n
    else d = power(a, b-1, n) # odd
        return (a*d) mod n
