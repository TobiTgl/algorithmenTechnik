# Türme Problem 
(Gleich wie Damenproblem nur threat ohne diagonale)
Es sei ein quadratisches Schachbrett mit den Dimensionen nXn gegeben. Bestimmen
Sie alle Möglichkeiten, wie man n Türme so aufstellen kann, dass sie sich nicht gegenseitig schlagen.


int main (){
    int n,i,noSol = 0;
    vector x; x.push_back(-1);      
    while (!x.empty()){
        int i = x.size() -1; 
        bool valid = false 
        while (!valid && x[i]<n-1) { //solange !valid und nicht letzte Spalte
            x[i]++; valid = true;
            for (j=0; j<i; j++) if (threat(x,i,j)) valid = false
        }
        if (valid) 
            if (i==n-1) { // Lösung gefunden
                writeSolution(x); noSol++
            } else x.push_back(-1);      // Vorwärts suchen (adde neue Zeile)
        } else x.pop_back               // Rückwärts suchen (entferne element == gehe Zeile zurück)
    }
return 0
}

bool threat(vector x, int j, int i){
    // Wird Feld x[j] durch x[i]
    return (x[i]==[j]);
}

T(n) = n^n

    }
return 0

bool threat(vector x, int j, int i){
    // Wird Feld x[j] durch x[i]
    return (x[i]==[j] || abs(x[i]-x[j])==j-i);
}

# Partitionen einer natürlichen Zahl



void partition(int n){
    int k,no;
    int x[500],s[500];
    bool isSucc;
    k = 1;              länge von x[]
    no = 0;
    x[k] = 0;
    s[0] = 0;
    
    while (k>0) {
    // Geeigneten Summanden finden
    do {
        isSucc = x[k]<n;
        if ( isSucc ) {
            x[k]++;
            s[k] = s[k-1]+x[k];
        }
    } while ( isSucc && !isCandidate(s[k]<=n) );
    // Summanden und Summe auswerten
    if (isSucc) { // -> isCandidate(s[k],n)== true
        if (s[k]==n) { // Summe hat Wert n erricht
            writeSolution(x,k);
            no++;
        } else { // -> Summe n noch nicht erreicht
            k++; // -> Füge weiteren Summanden hinzu
            x[k]=x[k-1]-1; // Dieser ist mindestens so…
            // …groß wie der letzte Summand.
        }
    } else { // x[k] ist kein gültiger Nachfolger
        k--; // -> Gehe einen Summanden zurück
    }
}
cout<<"Anzahl der Loesungen: " << no << endl;

void writeSolution(const vector<int>& x, ofstream &out){
    out << x[0];
    for(int i=1; i<x.size(); i++) out << " + " << x[i];
    out << endl;
}
