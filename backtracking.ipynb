def binpacking(bins, packages):
        
    vector = []             // bin assignments for packages vector speichert in welchem bin paket mit index landet
    solutions = []
    vector.append(-1)

    while (len(vector) > 0):
        package_index = len(vector) - 1     //startet bei 0
        valid = False

        while not valid and vector[package_index] < len(bins) - 1:  //endet wenn alle bins befüllt oder  letztes paket nicht reinpasst vector[package_index] < len(bins) - 1 vector an der stelle (paket) muss kleiner sein als len bins(in welchem bin gespeichert) 
            vector[package_index] += 1                              // erstmal werden alle rucksäcke befüllt
            valid = True
            
            if bins[vector[package_index]] < packages[package_index]:   // Gegenstand passt nicht rein
                valid = False

        if valid:
            bins[vector[package_index]] -= packages[package_index]  //Größe rucksack - paket
            if len(vector) == len(packages):                        //alle pakete Verteilt
                solutions.append(vector.copy())   //Füge lösung hinzu
            else:
                vector.append(-1)                 //Weiter zum nächsten Rucksack

        else:
            vector.pop()                        //Keine lösung, schritt zurrück

    return solutions

```python
V = 4           #number of nodes
answer = []
 
# Function to find the minimum weight 
# Hamiltonian Cycle
def tsp(graph, v, currPos, n, count, cost):
 
    # If last node is reached and it has 
    # a link to the starting node i.e 
    # the source then keep the minimum 
    # value out of the total cost of 
    # traversal and "ans"
    # Finally return to check for 
    # more possible values
    if (count == n and graph[currPos][0]):      #wenn alle Knoten besucht wurden und der letzte Knoten eine Verbindung zum Startknoten hat
        answer.append(cost + graph[currPos][0])
        return
 
    # BACKTRACKING STEP
    # Loop to traverse the adjacency list
    # of currPos node and increasing the count
    # by 1 and cost by graph[currPos][i] value
    for i in range(n):
        if (v[i] == False and graph[currPos][i]):   # verbindung zum nächsten Knoten besteht und der Knoten noch nicht besucht wurde	
            # Mark as visited
            v[i] = True
            tsp(graph, v, i, n, count + 1, cost + graph[currPos][i])
             
            # Mark ith node as unvisited
            v[i] = False
 
# Driver code
 
# n is the number of nodes i.e. V
if __name__ == '__main__':
    n = 4
    graph= [[ 0, 10, 15, 20 ],      # von 0 nach 0, 1, 2, 3
            [ 10, 0, 35, 25 ],      # von 1 nach 0, 1, 2, 3
            [ 15, 35, 0, 30 ],      # von 2 nach 0, 1, 2, 3
            [ 20, 25, 30, 0 ]]      # von 3 nach 0, 1, 2, 3
 
    # Boolean array to check if a node
    # has been visited or not
    v = [False for i in range(n)]
     
    # Mark 0th node as visited
    v[0] = True
 
    # Find the minimum weight Hamiltonian Cycle
    tsp(graph, v, 0, n, 1, 0)#graph, v, currPos, n, count, cost
 
    # ans is the minimum weight Hamiltonian Cycle
    print(min(answer))
 
# This code is contributed by mohit kumar
```
