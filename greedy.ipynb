# Approximate bin packing

wie viele Kisten werden benötigt, um Gegenstände zu verpacken
Gegenstände haben unterschiedliche Größen (< 1)
Kisten haben Größe 1
Greedy: 
jeder Gegenstand wird in die erste Kiste gepackt, in die er passt
wenn er in keine passt, wird eine neue Kiste geöffnet

function approximate_bin_packing(ItemSizes[]):
    bins = []
    for (i = 0; i < len(ItemSizes); i++):       // Iteriere über jedes Item
        for (j = 0; j < len(bins); j++):        // Iteriere über alle vorhandenen Bins
            if (ItemSizes[i] <= 1 - bins[j]):   // Versuche das Item in den Bin zu packen (größe in dem Beispiel 1)
                bins[j] += ItemSizes[i]         // Füge das Item zum Bin hinzu (Greedy-Schritt)
                break
        if (j == len(bins)):
            bins.append(ItemSizes[i])           // Wenn kein passender Bin gefunden, erstelle neuen Bin
    return len(bins)                            // Anzahl der verwendeten Bins zurückgeben

def partition(Items[]):
    Items.sort()                                    # -> O(n log n) Ähnlich wie bei franktionalen Rucksackproblem
    left = []; left_sum = 0
    right = []; right_sum = 0
    for item in Items:                              # -> O(n)
        if left_sum < right_sum:
            left.append(item); left_sum += item
        else:
            right.append(item); right_sum += item
    return (left, right)

# Set cover problem
Given a set of elements {1, 2, …, n} (called the universe) and a collection S of m subsets whose union equals the universe, the set cover problem is to identify the smallest sub-collection of S whose union equals the universe.

 U = {1, 2, 3, 4, 5} and the collection of sets S = { {1, 2, 3}, {2, 4}, {3, 4}, {4, 5} }. 
 Clearly the union of S is U. However, we can cover all elements with only two sets: { {1, 2, 3}, {4, 5} }, see picture. Therefore, the solution to the set cover problem has size 2.

def color_map(NeighborMatrix[][]):
    n = len(NeighborMatrix)                 // Determines the number of nodes in the graph, assumed to be 
    colors = [-1] * n                       // Initializes an array colors of size n where each node starts with color -1 (unassigned).

    for i in n:                                         # -> O(n)
        valid_color = False
        color = -1                          //  Initializes an array colors of size n where each node starts with color -1 (unassigned).
        while not valid_color:                          # -> O(c)
            color += 1                                  # Greedy-Schritt: Farbe, die noch kein Nachbar hat
            valid_color = True
            for j in n:                                 # -> O(n)
                if NeighborMatrix[i][j] == 1 and colors[j] == color:  // if already colored or color i same break loop
                    valid_color = False
                    break
        colors[i] = color                  // Once a valid color is found for node i, assign color to colors[i]

    return max(colors) + 1                 //  After all nodes are colored, determine the number of distinct colors used by taking the maximum value 

T(n)=O(n^2 * c)

# Flüsse in Netzwerken (Ford Fulkerson)
computes the maximum flow in a flow network (Graph mit Kanten und kapazität pro kante)
as long as there is a path from the source (start node) to the sink (end node), with available capacity on all edges in the path, we send flow along one of the paths. Then we find another path, and so on. A path with available capacity is called an augmenting path.

Start with initial flow as 0.
While exists augmenting path source to sink:  
    Find augmenting path using any path-finding algorithm
    Determine the amount of flow that can be sent along the augmenting path, which is the minimum residual capacity along the edges of the path.
    Increase the flow along the augmenting path by the determined amount.
    Return the maximum flow.


T(n)=O(maximum flow * Anzahl der Kanten(*n log n?)) // Es werden alle Kanten durchgesucht um pfad zu finden
                                         // iteratively increase the maximum flow to get the solution (worst case add 1 flow unit per iteration)
def ford_fulkerson(graph, source, sink):
    flow = 0
    while True:
        path = path_finding_algorithm(graph, source, sink)
        if not path:
            break
        min_capacity = min_capacity_of_path(path)
        flow += min_capacity
    return flow
