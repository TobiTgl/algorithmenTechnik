# Approximate bin packuing

wie viele Kisten werden benötigt, um Gegenstände zu verpacken
Gegenstände haben unterschiedliche Größen (< 1)
Kisten haben Größe 1
Greedy: 
jeder Gegenstand wird in die erste Kiste gepackt, in die er passt
wenn er in keine passt, wird eine neue Kiste geöffnet

function approximate_bin_packing(ItemSizes[]):
    bins = []
    for (i = 0; i < len(ItemSizes); i++):       // Iteriere über jedes Item
        for (j = 0; j < len(bins); j++):        // Iteriere über alle vorhandenen Bins
            if (ItemSizes[i] <= 1 - bins[j]):   // Versuche das Item in den Bin zu packen (größe in dem Beispiel 1)
                bins[j] += ItemSizes[i]         // Füge das Item zum Bin hinzu (Greedy-Schritt)
                break
        if (j == len(bins)):
            bins.append(ItemSizes[i])           // Wenn kein passender Bin gefunden, erstelle neuen Bin
    return len(bins)                            // Anzahl der verwendeten Bins zurückgeben

def partition(Items[]):
    Items.sort()                                    # -> O(n log n) Ähnlich wie bei franktionalen Rucksackproblem
    left = []; left_sum = 0
    right = []; right_sum = 0
    for item in Items:                              # -> O(n)
        if left_sum < right_sum:
            left.append(item); left_sum += item
        else:
            right.append(item); right_sum += item
    return (left, right)

T(N)= O(n log n)

# Set cover problem
Given a set of elements {1, 2, …, n} (called the universe) and a collection S of m subsets whose union equals the universe, the set cover problem is to identify the smallest sub-collection of S whose union equals the universe.

 U = {1, 2, 3, 4, 5} and the collection of sets S = { {1, 2, 3}, {2, 4}, {3, 4}, {4, 5} }. 
 Clearly the union of S is U. However, we can cover all elements with only two sets: { {1, 2, 3}, {4, 5} }, see picture. Therefore, the solution to the set cover problem has size 2.

function setCover(Universe, Sets){
    U0 = Universe
    C = leereMenge
    i=0
    while Ui != leereMenge                      // O(Universe)
        select set s from Sets that maxmizes S ∩ U  // O(s)  greedy step choosing a subset S that covers as many uncovered elements as possible
        Ui+1 = Ui - S                               // updates the set of remaining uncovered elements
        C = C ∪ {S}                                 // places S into C
        i++
    
    return C
}
